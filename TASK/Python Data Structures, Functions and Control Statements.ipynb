```python

#LIST
# A List is an ordered and mutable collection.
# It allows duplicate values and uses square brackets [].
print("----- LIST -----")
students = ["Devika", "Priya", "Anu"]
print(students)
print("First Student:", students[0])
students.append("Kiran")
print("After Append:", students)
students.insert(1, "Rahul")
print("After Insert:", students)
students.remove("Anu")
print("After Remove:", students)
students.sort()
print("Sorted:", students)
print("Length:", len(students))
```

```python

#TUPLE
# A Tuple is an ordered but immutable collection.
# Once created, elements cannot be modified.
print("\n----- TUPLE -----")
days = ("Mon", "Tue", "Wed", "Thu", "Fri")
print(days)
print("Second Day:", days[1])
print("Length:", len(days))
print("Index of Wed:", days.index("Wed"))
```

```python
# SET
# A Set is an unordered collection of unique elements.
# Duplicate values are automatically removed.
print("\n----- SET -----")
A = {10, 20, 30, 40}
B = {30, 40, 50, 60}
print("Set A:", A)
print("Set B:", B)
print("Union:", A.union(B))
print("Intersection:", A.intersection(B))
print("Difference:", A.difference(B))
A.add(100)
print("After Add:", A)
```

```python
# DICTIONARY
# Dictionary stores data in key-value pairs.
print("\n----- DICTIONARY -----")
employee = {
    "id": 101,
    "name": "Devika",
    "salary": 50000
}
print(employee)
print("Keys:", employee.keys())
print("Values:", employee.values())
print("Items:", employee.items())
employee.update({"salary": 60000})
print("Updated Dictionary:", employee)
```

```python

#STRINGS
# String is a sequence of characters enclosed in quotes.
print("\n----- STRINGS -----")
text = "python programming"
print(text.upper())
print(text.lower())
print(text.capitalize())
print(text.title())
print(text.replace("python", "Java"))
print("Length:", len(text))
print("Substring:", text[0:6])
```

```python
# FUNCTIONS
# Functions are reusable blocks of code.
# They help reduce repetition.
print("\n----- FUNCTIONS -----")
def add(a, b):
    return a + b
result = add(10, 20)
print("Sum =", result)
```

```python
# IF STATEMENT
# If statement executes a block only when condition is True.
print("\n----- IF -----")
age = 20
if age >= 18:
    print("Eligible for Voting")

# IF ELSE
# If-Else allows choosing between two alternatives.
print("\n----- IF ELSE -----")
number = 7
if number % 2 == 0:
    print("Even")
else:
    print("Odd")


# IF ELIF ELSE
# Used when multiple conditions need to be checked.
print("\n----- IF ELIF ELSE -----")
marks = 75
if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python

# NESTED IF
# An If statement inside another If statement.
print("\n----- NESTED IF -----")
age = 22
citizen = True
if citizen:
    if age >= 18:
        print("Can Vote")
    else:
        print("Age Not Sufficient")
```

```python

# FOR LOOP
# For loop is used to iterate over a sequence.
print("\n----- FOR LOOP -----")
for i in range(1, 6):
    print(i)


#WHILE LOOP
# While loop executes as long as condition remains True.
print("\n----- WHILE LOOP -----")
num = 1
while num <= 5:
    print(num)
    num += 1
```

```python

# PRIME NUMBER
# A Prime Number has exactly two factors: 1 and itself.
print("\n----- PRIME NUMBER -----")
n = 13
count = 0
for i in range(1, n + 1):
    if n % i == 0:
        count += 1
if count == 2:
    print(n, "is Prime")
else:
    print(n, "is Not Prime")
```

```python

# FACTORIAL
print("\n----- FACTORIAL -----")
num = 5
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial =", fact)


# PALINDROME NUMBER
# A Palindrome reads the same forward and backward.
print("\n----- PALINDROME -----")
num = 121
temp = num
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10
if temp == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")

# ARMSTRONG NUMBER
# Sum of digits raised to the power of number of digits
# equals the original number.
print("\n----- ARMSTRONG -----")
num = 153
temp = num
power = len(str(num))
total = 0
while num > 0:
    digit = num % 10
    total += digit ** power
    num //= 10
if temp == total:
    print("Armstrong")
else:
    print("Not Armstrong")
```

```python

#PATTERN - STAR TRIANGLE
print("\n----- PATTERN 1 -----")
for i in range(1, 6):
    for j in range(i):
        print("*", end=" ")
    print()

# PATTERN - NUMBER TRIANGLE
print("\n----- PATTERN 2 -----")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()


#PATTERN - REVERSE STAR
print("\n----- PATTERN 3 -----")
for i in range(5, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()


# FUNCTION WITH DEFAULT ARGUMENT
# Default arguments are used when no value is provided.
print("\n----- FUNCTION DEFAULT ARGUMENT -----")
def greet(name="Student"):
    print("Welcome", name)
greet()
greet("Devika")
```
