Matplotlib is one of the most popular Python libraries used for data visualization. It helps us represent data graphically so that trends, patterns, and comparisons become easier to understand.

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [120, 150, 180, 170, 210]
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(["Sales"])
plt.show()
```

```python
import matplotlib.pyplot as plt

attendance = [95, 90, 85, 80, 75, 70]
bonus_marks = [10, 8, 6, 4, 2, 0]

plt.plot(attendance, bonus_marks, marker='*')
plt.xlabel("Attendance (%)")
plt.ylabel("Bonus Marks")
plt.title("Attendance vs Bonus Marks")
plt.legend(["Bonus"])
plt.show()
```

Problem : Student Marks Analysis

Create a dataset containing marks of 10 students.

students = ['A','B','C','D','E','F','G','H','I','J']

marks = [78,85,92,67,88,95,72,81,90,76]

Tasks

Display student marks graphically.

Find the highest scorer.

Add appropriate title and axis labels.

Display marks in descending order and plot again.

```python
import matplotlib.pyplot as plt

# Dataset
students = ['A','B','C','D','E','F','G','H','I','J']
marks = [78,85,92,67,88,95,72,81,90,76]

# Display student marks graphically
plt.figure(figsize=(8,4))
plt.bar(students, marks)
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
# Find highest scorer
highest_mark = max(marks)
highest_student = students[marks.index(highest_mark)]

print("Highest Scorer:", highest_student)
print("Highest Marks:", highest_mark)
```

```python
# Display marks in descending order
sorted_data = sorted(zip(marks, students), reverse=True)

sorted_marks = [m for m, s in sorted_data]
sorted_students = [s for m, s in sorted_data]
print("\nMarks in Descending Order:")
for s, m in zip(sorted_students, sorted_marks):
    print(s, ":", m)
```

```python
# Plot sorted marks
plt.figure(figsize=(8,4))
plt.bar(sorted_students, sorted_marks)
plt.title("Student Marks (Descending Order)")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
