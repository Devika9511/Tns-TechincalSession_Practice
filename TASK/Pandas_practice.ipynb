```python
import pandas as pd
import numpy as np

# Create DataFrame
data = {
    'Employee': ['Ravi', 'Anita', 'John', 'Meena', 'Arun'],
    'Age': [25, 30, np.nan, 28, 35],
    'Salary': [30000, 45000, 40000, np.nan, 50000],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR']
}

df = pd.DataFrame(data)
print("Complete DataFrame:")
print(df)

print("First 5 Rows (head):")
print(df.head())


print("Last 5 Rows (tail):")
print(df.tail())
```

```python

print("Dataset Info:")
df.info()

print("Summary Statistics:")
print(df.describe())

print("Missing Values:")
print(df.isnull().sum())

# Fill missing values with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

print("After Filling Missing Values:")
print(df)

# Filter employees with salary > 40000
print("Employees with Salary > 40000:")
print(df[df['Salary'] > 40000])

# Add Bonus column
df['Bonus'] = df['Salary'] * 0.10

print("Data with Bonus:")
print(df)

# Group by Department
print("Average Salary by Department:")
print(df.groupby('Department')['Salary'].mean())
```
